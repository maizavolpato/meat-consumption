```python
#first let's import all necessery libraries for this analysis
import pandas as pd
import numpy as np

#using pandas library and 'read_csv' function to read amazon.csv file
df = pd.read_csv('meat_consumption.csv')
df.head()
```

```python
df.rename(columns = {'LOCATION':'Country', 'SUBJECT': 'Type_meat', 'MEASURE': 'Measure','TIME':'Year', 'Value': 'Total'}, inplace = True)
df.head()
```

```python
#Selecting only row with kg per capita and del years before 2000 and after 2017, becuse they're missing data
dfKgCap = df.loc[(df['Measure'] == 'KG_CAP') & (df['Year'] >= 2000) & (df['Year'] <= 2017)]
dfKgCap
```

```python
total_kg_year = dfKgCap.groupby(by = ['Year'])['Total'].sum().round(3).reset_index()
total_kg_year
```

```python
total_kg_country = dfKgCap.groupby(by = ['Country'])['Total'].sum().round(3).sort_values(ascending = False).reset_index()
total_kg_country
```

```python
dfUSA = dfKgCap.loc[(df['Country'] == 'USA')]
dfUSA = dfUSA.drop(['Country', 'Measure'], axis=1, inplace=False)
dfUSA 
```

```python
USA_per_year = dfUSA.groupby(by = ['Year'])['Total'].sum().round(3).reset_index()
USA_per_year 
```

```python
USA_mean = USA_per_year['Total'].mean().round(3)
USA_mean
```

```python
USA_type_meat = dfUSA.pivot(index='Year', columns='Type_meat', values='Total').round(3)
USA_type_meat 
```

```python
USA_type_meat_max = USA_type_meat.max()
USA_type_meat_max
```

```python
USA_type_meat_mean = USA_type_meat.mean().round(3)
USA_type_meat_mean
```

```python
dfAUS = dfKgCap.loc[(df['Country'] == 'AUS')]
dfAUS = dfAUS.drop(['Country', 'Measure'], axis=1, inplace=False)
dfAUS
```

```python
AUS_per_year = dfAUS.groupby(by = ['Year'])['Total'].sum().round(3).reset_index()
AUS_per_year 
```

```python
AUS_type_meat = dfAUS.pivot(index='Year', columns='Type_meat', values='Total').round(3)
AUS_type_meat
```

```python
dfISR = dfKgCap.loc[(df['Country'] == 'ISR')]
dfISR = dfISR.drop(['Country', 'Measure'], axis=1, inplace=False)
dfISR
```

```python
ISR_per_year = dfISR.groupby(by = ['Year'])['Total'].sum().round(3).reset_index()
ISR_per_year
```

```python
ISR_type_meat = dfISR.pivot(index='Year', columns='Type_meat', values='Total').round(3)
ISR_type_meat
```
